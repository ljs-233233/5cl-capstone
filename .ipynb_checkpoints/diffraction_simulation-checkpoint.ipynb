# Physics 5CL Capstone
The code in this notebook simulates the diffraction pattern of any pattern using the Fraunhofer Diffraction Equation. 

```python
# Import libraries
import numpy as np
import numpy.fft as fft
import matplotlib.pyplot as plt
```

```python
# Create a class that stores the grating shape and the diffraction pattern
class Diffraction:
    
    # Parameters: Shape of the grating as an array, size of canvas, etc
    def __init__(self, grating):
        assert grating.shape[0] == grating.shape[1], "The shape of the grating array should be a square."
        self.grating = grating
        self.shape = grating.shape
        self.pattern = self.fourier_transform()
        
    # Helper Function, calculate the diffraction pattern as a 2D array
    def fourier_transform(self):
        ft_real = np.absolute(fft.fft2(1 - self.grating))
        intensity = ft_real / np.sqrt(np.sum(ft_real ** 2))
        pattern = np.zeros(intensity.shape)
        for i in np.arange(self.shape[0]):
            for j in np.arange(self.shape[1]):
                pattern[i][j] = intensity[(i + self.shape[0] // 2) % self.shape[0]][(j + self.shape[1] // 2) % self.shape[1]]
        return pattern
    
    # Return the size of the canvas
    def get_size(self):
        return self.shape[0]
    
    # Return the grating shape array
    def get_grating(self):
        return self.grating
    
    # Return the diffraction pattern
    def get_diffraction(self):
        return self.pattern
```

```python
# Function that plots the grating shape and diffraction pattern
def plot(diffraction, lim, name, imax):
    center = diffraction.get_size() // 2
    
    fig = plt.figure()
    fig.set_size_inches(8, 8)
    ax = fig.add_subplot(111)
    ax.imshow(diffraction.get_grating(), cmap="gray")
    plt.show()
    
    fig = plt.figure()
    fig.set_size_inches(8, 8)
    ax = fig.add_subplot(111)
    ax.set_xlim(center - lim, center + lim)
    ax.set_ylim(center - lim, center + lim)
    ax.imshow(diffraction.get_diffraction(), cmap="gray", interpolation="bilinear", aspect='auto', vmax=imax)
    plt.show()
    
    return
```

```python
# Circle
circle_canvas = np.zeros((4096, 4096))
for i in np.arange(4096):
    for j in np.arange(4096):
        di = np.abs(i - 2048)
        dj = np.abs(j - 2048)
        d = np.sqrt(di ** 2 + dj ** 2)
        if d <= 256:
            circle_canvas[i][j] = 1

circle_name = "Circle"
circle = Diffraction(circle_canvas)
plot(circle, 256, circle_name, 0.001)
```

```python
# Cross
cross_canvas = np.zeros((4096, 4096))
for i in np.arange(4096):
    for j in np.arange(4096):
        if 2048 - 256 <= i <= 2048 + 256:
            if 2048 - 64 <= j <= 2048 + 64:
                cross_canvas[i][j] = 1

        if 2048 - 256 <= j <= 2048 + 256:
            if 2048 - 64 <= i <= 2048 + 64:
                cross_canvas[i][j] = 1

cross_name = "Cross"
cross = Diffraction(cross_canvas)
plot(cross, 128, cross_name, 0.005)
```
