```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Load the TIFF file
image_path = 'Target.tif'  # Path to your uploaded TIFF file
tiff_image = Image.open(image_path)
image_array = np.array(tiff_image)


# Display the image
plt.figure(figsize=(10, 8))  # You can adjust the size to your preference
plt.imshow(image_array, cmap='gray')  # 'gray' colormap for grayscale representation
plt.colorbar()  # Add a colorbar to show the intensity scale
plt.title('Top-Down View of TIFF File Using Grayscale Intensity')
plt.axis('off')  # Turn off the axis numbers and ticks
plt.show()
```
